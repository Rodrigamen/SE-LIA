Concatenamos todo el proceso investigado antes y vemos que funciona

```python
def extraer_top (celda):
    iter2 = celda.split(",")
    iter3 = iter2[0]
    iter4 = iter3.replace("<", "") and iter3.replace(">", "") and iter3.replace("{<", "")
    iter5 = iter4.split(":", 1)
    iter6 = iter5[1]
    iter7 = iter6.replace(">", "") and iter6.replace(" ", "") and iter6.replace("' '", "") 
    iter8 = iter7.split(":")
    iter9 = float(iter8[1])
    return iter9  
```

Procedemos a montar una función para cada rate a extraer por posición contando con sus variaciones para la limpieza dependiendo de la posición donde se encuentre el dato

```python
def extraer_jungla (celda):
    iter2 = celda.split(",")
    iter3 = iter2[1]
    iter4 = iter3.replace("<", "") and iter3.replace(">", "")
    iter5 = iter4.split(":", 1)
    iter6 = iter5[1]
    iter7 = iter6.replace(">", "") and iter6.replace(" ", "") and iter6.replace("' '", "") 
    iter8 = iter7.split(":")
    iter9 = float(iter8[1])
    return iter9  
```

```python
def extraer_mid (celda):
    iter2 = celda.split(",")
    iter3 = iter2[2]
    iter4 = iter3.replace("<", "") and iter3.replace(">", "")
    iter5 = iter4.split(":", 1)
    iter6 = iter5[1]
    iter7 = iter6.replace(">", "") and iter6.replace(" ", "") and iter6.replace("' '", "") 
    iter8 = iter7.split(":")
    iter9 = float(iter8[1])
    return iter9  
```

```python
def extraer_adc (celda):
    iter2 = celda.split(",")
    iter3 = iter2[3]
    iter4 = iter3.replace("<", "") and iter3.replace(">", "")
    iter5 = iter4.split(":", 1)
    iter6 = iter5[1]
    iter7 = iter6.replace(">", "") and iter6.replace(" ", "") and iter6.replace("' '", "") 
    iter8 = iter7.split(":")
    iter9 = float(iter8[1])
    return iter9  
```

```python
def extraer_support (celda):
    iter2 = celda.split(",")
    iter3 = iter2[4]
    iter4 = iter3.replace("<", "") and iter3.replace(">", "") and iter3.replace("}", "")
    iter5 = iter4.split(":", 1)
    iter6 = iter5[1]
    iter7 = iter6.replace(">", "") and iter6.replace(" ", "") and iter6.replace("' '", "") 
    iter8 = iter7.split(":")
    iter9 = float(iter8[1])
    return iter9  
```

```python
campeones["TOP Play Rate"] = campeones["Play Rate"].apply(extraer_top)
campeones["Jungle Play Rate"] = campeones["Play Rate"].apply(extraer_jungla)
campeones["MID Play Rate"] = campeones["Play Rate"].apply(extraer_mid)
campeones["ADC Play Rate"] = campeones["Play Rate"].apply(extraer_adc)
campeones["Support Play Rate"] = campeones["Play Rate"].apply(extraer_support)
```

Columna nueva para, ejecutando la siguiente función, asignar posición preferente a cada campeón

```python
def cutrez (data, col1, col2, col3, col4, col5, col6, val1, val2, val3, val4, val5):
  for i in data.index:
    if (data[col1][i]) > (data[col2][i]) and (data[col1][i]) > (data[col3][i]) and (data[col1][i]) > (data[col4][i]) and (data[col1][i]) > (data[col5][i]):
      data.loc[i, col6]=val1
    elif (data[col2][i]) > (data[col1][i]) and (data[col2][i]) > (data[col3][i]) and (data[col2][i]) > (data[col4][i]) and (data[col2][i]) > (data[col5][i]):
      data.loc[i, col6]=val2
    elif (data[col3][i]) > (data[col1][i]) and (data[col3][i]) > (data[col2][i]) and (data[col3][i]) > (data[col4][i]) and (data[col3][i]) > (data[col5][i]):
      data.loc[i, col6]=val3    
    elif (data[col4][i]) > (data[col1][i]) and (data[col4][i]) > (data[col2][i]) and (data[col4][i]) > (data[col3][i]) and (data[col4][i]) > (data[col5][i]):
      data.loc[i, col6]=val4
    elif (data[col5][i]) > (data[col1][i]) and (data[col5][i]) > (data[col2][i]) and (data[col5][i]) > (data[col3][i]) and (data[col5][i]) > (data[col4][i]):
      data.loc[i, col6]=val5
```

Función, por la cual pido perdón de antemano, que lo que hace es buscar, por campeón, en cuál posición tiene mayor Play Rate para saber a qué posición pertenece cada campeón de forma preferente

Insertamos columnas para poder almacenar en ellas el total de las estadísticas y el incremento total a nivel máximo de cada estadística

```python
def incremento (celda):
  valor = celda * 17
  return valor
```

Función para calcular el incremento. PD: Son 18 niveles y se empieza en el 1, por eso multiplicamos por 17

```python
campeones["Incremento HP"] = campeones["hp+"].apply(incremento)
campeones["Incremento Magic Power"] = campeones["mp+"].apply(incremento)
campeones["Incremento Attack Damage"] = campeones["ad+"].apply(incremento)
campeones["Incremento Attack Speed"] = campeones["as+"].apply(incremento)
campeones["Incremento Attack Resistance"] = campeones["ar+"].apply(incremento)
campeones["Incremento Magic Resistance"] = campeones["mr+"].apply(incremento)
```

```python
def totales (base, incremento):
  for ir in range (0, 145):
    total = base + incremento
    return total  
```
