```python
listanumeros = normalizado["Unnamed: 0"].tolist()
listaids = campeones["ID"].tolist()
```

```python
dictvacio = {}
```

```python
for i, z in zip(listanumeros, listaids):
  dictvacio[i] = z
```

```python
normalizado["ID"] = normalizado["Unnamed: 0"].map(dictvacio)
```
